```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style("ticks")
sns.set_style({"xtick.direction": "in","ytick.direction": "in"})
sns.color_palette("colorblind")
```

```python
def free_cont_1d(omega):
    result = 1/np.sqrt(omega)
    bool_arr = omega <= 0
    result[bool_arr] = 0
    return result

def sommerfeld_enh(omega):
    
def free_cont(omega):
    result = np.sqrt(omega)
    bool_arr = omega < 0
    result[bool_arr] = 0
    return result

def sommerfeld_enh(omega):
    pi_omega = np.pi/np.sqrt(omega)
    result = pi_omega * np.exp(pi_omega)/np.sinh(pi_omega)
    
    bool_arr = omega <= 0.0
    result[bool_arr] = 0
    return result

def gauss(omega):
    result = np.exp(-(omega)**2/0.1)
    result[omega > 0] = 0
    return result

def gauss_ul(omega, upper, lower):
    result = upper*np.exp(-(omega)**2/0.1)
    result[omega > 0] = lower*np.exp(-(omega[omega>0])**2/0.1)
    return result
```

```python
omega = np.linspace(-2, 10, 1000)
offset = np.linspace(-2, 10, 1000)
#offset[omega < 0] = 0
#offset[omega >= 0] = -6.4
```

```python
plt.plot(omega, 1.8*free_cont(omega), "--", label="No Coulomb Interactions")
plt.plot(omega, free_cont(omega)*sommerfeld_enh(omega) + gauss_ul(omega, 7,0.71), label="Coulomb-enhanced Absorption")
plt.xlabel("$\hbar\omega - E_{g}$ (arbitrary units)")
plt.ylabel("Absorption (arbitrary units)")
plt.xlim([-1, 10])

plt.ylim([0,8])
plt.legend()
plt.savefig("coulomb_enhancement.pdf")
```
